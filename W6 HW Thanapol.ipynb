Question 1:

Simple linear regression model is a linear graph use to estimate and understand the relationship between two quantitative variables (x,y) (independent and dependent variables). We want to find out that the the predictor variable,X is the factor we believe explains or influences changes in changes in Y. The model includes two main coefficients: the slope coefficient (ùõΩ1) and the intercept (ùõΩ0). The slope coefficient, represents the rate at which y changes for a one-unit increase in x indicating a positive or negative relationship based on its sign. The intercept, is the starting point of the regression line which is the value of y when x is zero. An error term, œµ, accounts for variations in y that are not explained by x, representing random noise or other unmeasured factors. Together, these components form the Simple Linear Regression model equation. Assuming the error term follows a normal distribution with a mean of zero, this model implies that for any fixed value of x , the outcome y is normally distributed around the predicted value ùõΩ0+ùõΩ1ùëã or the linear line, with deviations due to random error.

Thus, Simple Linear Regression provides a way to model and predict y as a linear function of x, with observations clustering around a line defined by ùõΩ0 and ùõΩ1, within a range of normally distributed variation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the theoretical model
n = 100                    # Number of samples
beta0 = 2.0                # Intercept
beta1 = 0.5                # Slope
sigma = 1.0                # Standard deviation of the error term

# Generate predictor values (x) uniformly within a specified range
x = np.linspace(0, 10, n)

# Generate random errors (epsilon) from a normal distribution with mean 0 and standard deviation sigma
errors = norm.rvs(loc=0, scale=sigma, size=n)

# Calculate the outcome variable (Y) based on the linear model Y = beta0 + beta1 * X + epsilon
y = beta0 + beta1 * x + errors

# Plotting the data to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.6, label="Observed Data (Y)")
plt.plot(x, beta0 + beta1 * x, color='red', label="True Regression Line (without error)")
plt.xlabel("Predictor Variable (X)")
plt.ylabel("Outcome Variable (Y)")
plt.title("Simple Linear Regression Model with Random Error")
plt.legend()
plt.show()
```

The "true line" (green): is the "ideal" regression line based on known parameters, which represents the perfect linear relationship without any random error which is very unlikely to happen in real life because all observed data points contain some level of random variation or error. 

Fitted Line (Red): is based on the data that includes random error, the model uses the sample to estimate the best-fitting line. Becuase of the impact of random error, each fitted line would differ slightly if we repeated this process with new samples.

purpose

The purpose of comparing the true and fitted lines is to show the impact of random sampling variation.
```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Parameters for the theoretical model
n = 100                    # Number of samples
beta0 = 2.0                # Intercept
beta1 = 0.5                # Slope
sigma = 1.0                # Standard deviation of the error term

# Generate predictor values (x) and random errors
x = np.linspace(0, 10, n)
errors = np.random.normal(0, sigma, n)

# Calculate outcome variable (y) based on the linear model
y = beta0 + beta1 * x + errors

# Create a DataFrame to store the simulated data
data = pd.DataFrame({'X': x, 'Y': y})

# Fit the Simple Linear Regression model using statsmodels
model = smf.ols('Y ~ X', data=data).fit()

# Print the summary of the model to see estimated coefficients and statistics
print(model.summary())

# Visualize both the true regression line and the fitted regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='blue', alpha=0.6, label="Observed Data (Y)")
plt.plot(data['X'], beta0 + beta1 * data['X'], color='green', label="True Regression Line (without error)")
plt.plot(data['X'], model.predict(data['X']), color='red', label="Fitted Regression Line")
plt.xlabel("Predictor Variable (X)")
plt.ylabel("Outcome Variable (Y)")
plt.title("Comparison of True and Fitted Simple Linear Regression Lines")
plt.legend()
fig.show(renderer="png")
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import matplotlib.pyplot as plt

# Define the long_wait_limit value
long_wait_limit = 71

# Load your dataset (assuming it's already available in the 'old_faithful' DataFrame)
# Filter for long waiting times
long_wait = old_faithful[old_faithful['waiting'] > long_wait_limit]

# Step 1: Bootstrap Simple Linear Regression

# Function to perform bootstrap and collect slope coefficients
def bootstrap_slope(data, n_iterations=1000):
    slopes = []
    for _ in range(n_iterations):
        sample = data.sample(n=len(data), replace=True)  # Sampling with replacement
        model = smf.ols('duration ~ waiting', data=sample).fit()  # Fit the model
        slopes.append(model.params['waiting'])  # Store the slope coefficient
    return np.array(slopes)

# Perform bootstrapping
bootstrap_slopes = bootstrap_slope(long_wait)

# Visualize the distribution of the slope coefficients from the bootstrap samples
plt.hist(bootstrap_slopes, bins=30, edgecolor='black')
plt.axvline(x=0, color='red', linestyle='--', label="Null Hypothesis (B1=0)")
plt.xlabel("Slope Coefficients")
plt.ylabel("Frequency")
plt.title("Bootstrapped Sampling Distribution of the Slope Coefficients")
plt.legend()
plt.show()

# Step 2: Simulate Under Null Hypothesis (B1 = 0)

# Simulate data under the null hypothesis
def simulate_null_data(n, B0=1.65, B1=0, sigma=0.37):
    x = np.random.uniform(low=min(long_wait['waiting']), high=max(long_wait['waiting']), size=n)
    y = B0 + B1 * x + np.random.normal(0, sigma, n)
    return pd.DataFrame({'waiting': x, 'duration': y})

# Function to compute p-value under the null hypothesis
def compute_p_value(simulated_slopes, observed_slope):
    # p-value is the proportion of simulated slopes greater than or equal to the observed slope
    return np.mean(simulated_slopes >= observed_slope)

# Number of simulations
n_simulations = 1000

# Simulate the null hypothesis data
simulated_slopes = []
for _ in range(n_simulations):
    simulated_data = simulate_null_data(n=len(long_wait))
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    simulated_slopes.append(model.params['waiting'])

# Convert to a NumPy array
simulated_slopes = np.array(simulated_slopes)

# Fit the model to the original data and get the observed slope
observed_model = smf.ols('duration ~ waiting', data=long_wait).fit()
observed_slope = observed_model.params['waiting']

# Visualize the distribution of the simulated slopes
plt.hist(simulated_slopes, bins=30, edgecolor='black')
plt.axvline(x=observed_slope, color='blue', linestyle='--', label="Observed Slope")
plt.axvline(x=0, color='red', linestyle='--', label="Null Hypothesis (B1=0)")
plt.xlabel("Slope Coefficients")
plt.ylabel("Frequency")
plt.title("Simulated Sampling Distribution of the Slope Coefficients (Null Hypothesis)")
plt.legend()
plt.show()

# Compute p-value for the observed slope
p_value = compute_p_value(simulated_slopes, observed_slope)
print(f"P-value for the observed slope: {p_value}")
fig.show(renderer="png")
```

### (A) Solution for the Permutation Test

1. **Define Hypothesis**:
   - **Null Hypothesis (\( H_0 \))**: The mean wait times for the "short" and "long" groups are equal (\( \mu_{\text{short}} = \mu_{\text{long}} \)).

2. **Calculate Observed Mean Difference**:
   - Compute the actual difference in means between the "short" and "long" groups based on the original data. Let‚Äôs call this value \( D_{\text{obs}} \).

3. **Shuffle Labels and Compute New Mean Differences**:
   - Repeat the following steps many times (e.g., 10,000 iterations):
     - Shuffle the "short" and "long" labels randomly across all data points.
     - Split the data into new "short" and "long" groups based on the shuffled labels.
     - Calculate the difference in means for the shuffled groups.
   - This creates a distribution of mean differences under the assumption that there‚Äôs no true difference between the groups.

4. **Calculate p-Value**:
   - Determine the proportion of shuffled mean differences that are as extreme as or more extreme than the observed difference \( D_{\text{obs}} \).
   - If this p-value is small (e.g., \( p < 0.05 \)), we reject the null hypothesis and conclude that there‚Äôs a significant difference in mean wait times between the "short" and "long" groups.



### (B) Solution for the 95% Bootstrap Confidence Interval

1. **Generate Bootstrap Samples**:
   - For each group (short and long), create a large number of bootstrap samples by sampling with replacement. For instance, if each group has 50 observations, draw 50 observations with replacement from the "short" group and 50 from the "long" group.
   
2. **Calculate Mean Differences for Each Bootstrap Sample**:
   - For each bootstrap sample, compute the mean of the "short" sample and the mean of the "long" sample, then calculate the difference in means.
   - Repeat this process many times (e.g., 10,000 times) to create a distribution of bootstrap mean differences.

3. **Compute Confidence Interval**:
   - Use `np.quantile` to find the 2.5th and 97.5th percentiles of the bootstrap mean difference distribution.
   - These percentiles form the bounds of a 95% confidence interval for the mean difference.


### (a) Explanation of Sampling Approaches

**Permutation Test**:
- The permutation test tests the null hypothesis by treating the labels as if they were exchangeable. By shuffling the labels and recalculating the mean difference, we create a distribution of differences that would occur if there were no real difference between the groups.
- This test is particularly useful for hypothesis testing when we want to determine whether an observed difference is significant.

**Bootstrap Confidence Interval**:
- Bootstrapping estimates the variability of the mean difference by resampling with replacement within each group, building an empirical distribution of differences between the sample means.
- This method provides a confidence interval rather than a p-value, allowing us to quantify uncertainty in the estimated mean difference without making distributional assumptions.
